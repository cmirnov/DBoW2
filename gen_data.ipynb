```python
import matplotlib.pyplot as plt
import numpy as np


from mpl_toolkits import mplot3d

from mpl_toolkits.mplot3d import Axes3D
```

```python
branches = 9
depth = 3
num_leaves = 30
dim = 20
```

```python
def gen_data(centorid, level, points):
    if level == depth:
        for _ in range(branches):
            n_centroid = np.random.uniform(low=-1/(branches ** (level**2)), high=1/(branches ** (level**2)), size=[dim])
            n_centroid += centorid
            points += [n_centroid]
        return
    for _ in range(branches):
        n_centroid = np.random.uniform(low=-1/(branches ** (level**2)), high=1/(branches ** (level**2)), size=[dim])
        n_centroid += centorid
        gen_data(n_centroid, level + 1, points)
```

```python
np.random.uniform(low=0, high=10, size=[6])
```

```python
p = []
gen_data(0, 0, p)
p = np.array(p)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(p[:,0],p[:,1], p[:,2])
# plt.show()
```
